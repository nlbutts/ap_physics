```python
import matplotlib.pyplot as plt
import math
import numpy as np
plt.rcParams['figure.figsize'] = [10, 10]
```

# Equations
$$\alpha = \frac{dw}{dt} = \frac{d^2\omega}{dt^2}$$

If angular acceleration is constant:
$$w(t) = w0 + \alpha t$$
$$\omega(t) = \omega0 + w0t + 1/2\alpha t^2$$

Centripetal Acceleration
$$a_c = R\omega^2$$
$$\omega = \sqrt{a_c / R}$$

```python
accel_mss = 3
velocity_ms = 5
radius_meters = 300
t = 0
omega_ms = math.sqrt(accel_mss / radius_meters)
print(f'Calculated Angular Velocity: {omega_ms}')
print('Assumming constant angular velocity.')
```

```python
def calculate_pos_at_t(t):
    x = radius_meters * np.sin(-velocity_ms * t / radius_meters)
    y = -radius_meters * np.cos(-velocity_ms * t / radius_meters)
    return x,y
```

```python
# This loops through from 0 to 300 seconds with 25 total steps.
t = np.linspace(0, 300, 25)
x, y = calculate_pos_at_t(t)
```

```python
# This plots the scatter points and arrows from the previous sample to the next sample.
U = np.diff(x)
V = np.diff(y)
plt.scatter(x, y, label='points')
plt.quiver(x[:-1], y[:-1], U, V, angles='xy', scale_units='xy', scale=1, color='red')
plt.grid()
plt.title(f'Position of car with Ac={accel_mss} m/s^2 on a track of radius: {radius_meters} meters')
```
